```python
import matplotlib.pyplot as plt
from matplotlib import font_manager
import pandas as pd

# Matplotlib 设置中文字体
font_path = "/usr/share/fonts/truetype/wqy/wqy-microhei.ttc"  # Linux 示例路径
my_font = font_manager.FontProperties(fname=font_path)
plt.rcParams['font.family'] = my_font.get_name()  # 使用指定的中文字体

# 数据表 1: Performance Comparison with Lucid as Baseline
data1 = {
    '指标': ['平均作业完成时间', '平均队列长度', 'P99.9 队列长度'],
    'Lucid': [12886, 915, 93756],
    'w.o.打包分析': [13286, 1338, 109577],
    'w.o.吞吐量预测': [15098, 3387, 193383],
    'w.o.工作量估计': [16080, 4369, 221435]
}

# 数据表 2: Average JCT (hrs) Comparison with Lucid as Baseline
data2 = {
    '作业类型': ['大规模作业', '小规模作业'],
    'Lucid': [16539, 12469],
    'w.o.打包分析': [18389, 12704],
    'w.o.吞吐量预测': [24170, 14064],
    'w.o.工作量估计': [27462, 14012]
}

# 数据表 3: Average Queue (hrs) Comparison with Lucid as Baseline
data3 = {
    '作业类型': ['大规模作业', '小规模作业'],
    'Lucid': [3089, 667],
    'w.o.打包分析': [4939, 927],
    'w.o.吞吐量预测': [10721, 2551],
    'w.o.工作量估计': [14782, 3269]
}

# 绘制 Performance Comparison 子图并保存
def plot_performance_comparison(data, filename):
    metrics = ['平均作业完成时间', '平均队列长度', 'P99.9 队列长度']
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    df = pd.DataFrame(data)

    for i, metric in enumerate(metrics):
        values = df.iloc[i, 1:]  # 选取每一行数据
        labels = df.columns[1:]  # 获取列名 (不同配置)
        axes[i].bar(labels, values, color=['skyblue', 'orange', 'green', 'red'])
        axes[i].set_title(metric, fontproperties=my_font)
        axes[i].set_ylabel('时长（小时）', fontproperties=my_font)  # 设置纵坐标
        axes[i].set_xticks(range(len(labels)))
        axes[i].set_xticklabels(labels, rotation=45, fontproperties=my_font)

    plt.tight_layout()
    plt.savefig(filename, format='pdf')
    plt.show()

# 绘制 Average JCT 和 Average Queue 子图并保存
def plot_job_comparison(data, title, filename):
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    fig.suptitle(title, fontproperties=my_font, fontsize=16)  # 添加大标题
    df = pd.DataFrame(data)

    for i, job_type in enumerate(df['作业类型']):
        values = df.iloc[i, 1:]  # 选取每一行数据
        labels = df.columns[1:]  # 获取配置名称
        axes[i].bar(labels, values, color=['skyblue', 'orange', 'green', 'red'])
        axes[i].set_title(job_type, fontproperties=my_font)
        axes[i].set_ylabel('时长（小时）', fontproperties=my_font)  # 设置纵坐标
        axes[i].set_xticks(range(len(labels)))
        axes[i].set_xticklabels(labels, rotation=45, fontproperties=my_font)

    plt.tight_layout(rect=[0, 0, 1, 0.95])  # 调整布局以适应大标题
    plt.savefig(filename, format='pdf')
    plt.show()

# 调用绘图函数并保存
plot_performance_comparison(data1, "performance_comparison.pdf")
plot_job_comparison(data2, '平均作业完成时间（小时）对比', "average_jct_comparison.pdf")
plot_job_comparison(data3, '平均队列长度（小时）对比', "average_queue_comparison.pdf")
```
