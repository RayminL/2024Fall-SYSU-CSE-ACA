```python
import matplotlib.pyplot as plt
from matplotlib import font_manager
import pandas as pd

# Matplotlib 
font_path = "/usr/share/fonts/truetype/wqy/wqy-microhei.ttc"  # Linux 
my_font = font_manager.FontProperties(fname=font_path)
plt.rcParams['font.family'] = my_font.get_name()  # 

#  1: Performance Comparison with Lucid as Baseline
data1 = {
    '': ['', '', 'P99.9 '],
    'Lucid': [12886, 915, 93756],
    'w.o.': [13286, 1338, 109577],
    'w.o.': [15098, 3387, 193383],
    'w.o.': [16080, 4369, 221435]
}

#  2: Average JCT (hrs) Comparison with Lucid as Baseline
data2 = {
    '': ['', ''],
    'Lucid': [16539, 12469],
    'w.o.': [18389, 12704],
    'w.o.': [24170, 14064],
    'w.o.': [27462, 14012]
}

#  3: Average Queue (hrs) Comparison with Lucid as Baseline
data3 = {
    '': ['', ''],
    'Lucid': [3089, 667],
    'w.o.': [4939, 927],
    'w.o.': [10721, 2551],
    'w.o.': [14782, 3269]
}

#  Performance Comparison 
def plot_performance_comparison(data, filename):
    metrics = ['', '', 'P99.9 ']
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    df = pd.DataFrame(data)

    for i, metric in enumerate(metrics):
        values = df.iloc[i, 1:]  # 
        labels = df.columns[1:]  #  ()
        axes[i].bar(labels, values, color=['skyblue', 'orange', 'green', 'red'])
        axes[i].set_title(metric, fontproperties=my_font)
        axes[i].set_ylabel('', fontproperties=my_font)  # 
        axes[i].set_xticks(range(len(labels)))
        axes[i].set_xticklabels(labels, rotation=45, fontproperties=my_font)

    plt.tight_layout()
    plt.savefig(filename, format='pdf')
    plt.show()

#  Average JCT  Average Queue 
def plot_job_comparison(data, title, filename):
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    fig.suptitle(title, fontproperties=my_font, fontsize=16)  # 
    df = pd.DataFrame(data)

    for i, job_type in enumerate(df['']):
        values = df.iloc[i, 1:]  # 
        labels = df.columns[1:]  # 
        axes[i].bar(labels, values, color=['skyblue', 'orange', 'green', 'red'])
        axes[i].set_title(job_type, fontproperties=my_font)
        axes[i].set_ylabel('', fontproperties=my_font)  # 
        axes[i].set_xticks(range(len(labels)))
        axes[i].set_xticklabels(labels, rotation=45, fontproperties=my_font)

    plt.tight_layout(rect=[0, 0, 1, 0.95])  # 
    plt.savefig(filename, format='pdf')
    plt.show()

# 
plot_performance_comparison(data1, "performance_comparison.pdf")
plot_job_comparison(data2, '', "average_jct_comparison.pdf")
plot_job_comparison(data3, '', "average_queue_comparison.pdf")
```
